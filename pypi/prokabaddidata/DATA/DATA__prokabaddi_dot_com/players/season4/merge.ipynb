```python
import pandas as pd
import glob

# # Load all CSV files into a list of dataframes
# file_paths = [
#     "./1_df_season1_player-total-points.csv",
#     "./2_df_season1_player-successful-raids.csv",
#     "./3_df_season1_player-raid-points.csv",
#     "./4_df_season1_player-successful-tackles.csv",
#     "./5_df_season1_player-tackle-points.csv",
#     "./7_df_season1_player-avg-tackle-points.csv",
#     "./8_df_season1_player-dod-raid-points.csv",
#     "./9_df_season1_player-super-raids.csv",
#     "./10_df_season1_player-super-tackles.csv",
#     "./11_df_season1_super-10s.csv",
#     "./12_df_season1_high-5s.csv",
# ]

# # Reading all dataframes
# dfs = [pd.read_csv(file) for file in file_paths]

file_map = {
    "./1_df_season4_player-total-points.csv": "total_points",
    "./2_df_season4_player-successful-raids.csv": "successful_raids",
    "./3_df_season4_player-raid-points.csv": "raid_points",
    "./4_df_season4_player-successful-tackles.csv": "successful_tackles",
    "./5_df_season4_player-tackle-points.csv": "tackle_points",
    "./7_df_season4_player-avg-tackle-points.csv": "avg_tackle_points",
    "./8_df_season4_player-dod-raid-points.csv": "dod_raid_points",
    "./9_df_season4_player-super-raids.csv": "super_raids",
    "./10_df_season4_player-super-tackles.csv": "super_tackles",
    "./11_df_season4_super-10s.csv": "super_10s",
    "./12_df_season4_high-5s.csv": "super_5s",
}

# Load the CSVs again and rename columns based on their corresponding file
renamed_dfs = []

for path, stat_name in file_map.items():
    df = pd.read_csv(path)
    rename_dict = {
        col: (
            f"{stat_name}_{col}"
            if col not in ["player_id", "player_name", "team_name", "team_full_name"]
            else col
        )
        for col in df.columns
    }
    renamed_dfs.append(df.rename(columns=rename_dict))

# Merge all dataframes on 'player_id'
merged_df = renamed_dfs[0]
for df in renamed_dfs[1:]:
    merged_df = pd.merge(
        merged_df,
        df,
        on=["player_id", "player_name", "team_name", "team_full_name"],
        how="outer",
    )



# # Merge all dataframes on 'player+id'
# merged_df = dfs[0]
# for df in dfs[1:]:
#     merged_df = pd.merge(
#         merged_df, df, on="player_id", how="outer", suffixes=("", "_duplicate")
#     )

# Removing duplicate columns
merged_df_no_dup = merged_df.loc[:, ~merged_df.columns.duplicated()]

# tools.display_dataframe_to_user(name="Merged Player Statistics", dataframe=merged_df)
```
